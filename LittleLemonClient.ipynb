```python
import mysql.connector as connector
```

```python
### Task 1: Connect to the database

connection = connector.connect(user="root", password="password", db='littlelemondb')
cursor = connection.cursor()
```

```python
### Task 2: Show all tables in the database

show_tables_query = "SHOW tables"
cursor.execute(show_tables_query)

results = cursor.fetchall()
print(results)
```

```python
### Task 3: Query with table JOIN

cursor.execute(
    """
    SELECT customer.FullName, customer.ContactNumber
    FROM customer
    INNER JOIN orders ON customer.CustomerID = orders.CustomerID
    WHERE orders.TotalCost > 60;
    """
)

results = cursor.fetchall()

for result in results:
    print(result)
```
